# The overall sales trend

```python
df['Order Date'].min()
```

```python
df['Order Date'].max()
```

```python
df['MonthYear'] = df['Order Date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
# Grouping month year
df_trend = df.groupby('MonthYear')['Sales'].sum().reset_index()
```

```python
df_trend
```

```python
# setting the figure size
plt.figure(figsize=(15,6))

plt.plot(df_trend['MonthYear'], df_trend['Sales'])
plt.xticks(rotation='vertical',size=9)
plt.show()
```

# Top 10 products by sales

```python
product_sales = pd.DataFrame(df.groupby('Product Name')['Sales'].sum())
product_sales = product_sales.sort_values('Sales',ascending = False)
product_sales
```

```python
product_sales[:10]
```

# Most selling products

```python
# grouping product name
most_selling_product = pd.DataFrame(df.groupby('Product Name')['Quantity'].sum())
most_selling_product = most_selling_product.sort_values('Quantity',ascending=False)
most_selling_product[:10]
```

# Most preferred ship mode

```python
sns.countplot(df['Ship Mode'])
```

# Most profitable category and sub-category

```python
cat_subcat_profit = pd.DataFrame(df.groupby(['Category','Sub-Category'])['Profit'].sum())
cat_subcat_profit.sort_values(['Category','Sub-Category'],ascending=False)
```
