```python
movies['tags'] = movies['genre'] + movies['overview']
```

```python
movies.head()
```

```python
new_df = movies[['id','title','tags']]
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 10000, stop_words = "english")
```

```python
cv
```

```python
vec = cv.fit_transform(new_df['tags'].values.astype('U')).toarray()
```

```python
vec
```

```python
vec.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(vec)
```

```python
sim
```

```python
new_df[new_df['title'] == 'Dilwale Dulhania Le Jayenge']
```

```python
dist = sorted(list(enumerate(sim[1])), reverse = True, key = lambda vec:vec[1])
```

```python
dist
```

```python
for i in dist[0:5]:
    print(new_df.iloc[i[0]].title)
```

```python
def recommend(movies):
    index = new_df[new_df['title']==movies].index[0]
    distance = sorted(list(enumerate(sim[index])),reverse=True,key=lambda vec:vec[1])
    for i in distance[0:5]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend("Iron Man")
```
